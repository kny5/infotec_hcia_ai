```python
# 1 Import dependent packages

from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 2 Build and visualize a house price dataset.

x = np.array([121, 125, 131, 141, 152, 161]).reshape(-1,1)
y = np.array([300, 350, 425, 405, 496, 517])
```

```python
plt.scatter(x,y)
plt.xlabel("area")
plt.ylabel("price")
```

```python
# 3 Train the model

lr = LinearRegression()
lr.fit(x,y)
```

```python
# 4 Visualize the model
w = lr.coef_
b = lr.intercept_
print('slope:',w)
print('Intercept: ',b)
```

```python
plt.scatter(x,y)
plt.xlabel("area")
plt.ylabel("price")
plt.plot([x[0],x[-1]],[x[0]*w+b,x[-1]*w+b])
```

```python
# 5 Start prediction task using model
testX = np.array([[130]])
lr.predict(testX)
```
